```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
!pip install -U kaleido
```

```python
# Cargar los datos
url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
df = pd.read_json(url)
```

```python
# Columnas con diccionarios
columnas_dict = ['customer', 'phone', 'internet', 'account']
```

```python
# Transformar columnas con diccionarios a columnas simples
for col in columnas_dict:
    df_temp = pd.json_normalize(df[col])
    df = pd.concat([df, df_temp], axis=1)
    df.drop(columns=[col], inplace=True)
```

#üîß Transformaci√≥n

#### Comprobaci√≥n de incoherencias en los datos
##### Verificar si hay problemas en los datos que puedan afectar el an√°lisis. Prestar atenci√≥n a valores ausentes, duplicados, errores de formato e inconsistencias en las categor√≠as. Este proceso es esencial para asegurar que los datos est√©n listos para las siguientes etapas.
#### Estandarizaci√≥n y transformaci√≥n de datos
##### La estandarizaci√≥n y transformaci√≥n de datos es una etapa opcional, pero altamente recomendada, ya que busca hacer que la informaci√≥n sea m√°s consistente, comprensible y adecuada para el an√°lisis. Durante esta fase, por ejemplo, puedes convertir valores textuales como "S√≠" y "No" en valores binarios (1 y 0), lo que facilita el procesamiento matem√°tico y la aplicaci√≥n de modelos anal√≠ticos.

Adem√°s, traducir o renombrar columnas y datos hace que la informaci√≥n sea m√°s accesible y f√°cil de entender, especialmente cuando se trabaja con fuentes externas o t√©rminos t√©cnicos. Aunque no es un paso obligatorio, puede mejorar significativamente la claridad y comunicaci√≥n de los resultados, facilitando la interpretaci√≥n y evitando confusiones, especialmente al compartir informaci√≥n con stakeholders no t√©cnicos.

```python
# Crear la columna cuentas diarias dividiendo 'ChargesMonthly' por 30 y que el resultado muestre solo 2 decimales
df["Cuentas_Diarias"] = df["Charges.Monthly"] / 30
df["Cuentas_Diarias"] = df["Cuentas_Diarias"].round(2)
```

```python
df
```

```python
# Columnas que deben ser transformadas: s√≥lo las estrictamente "Yes"/"No"
bin_cols = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']

# Aplicar transformaci√≥n a binario
for col in bin_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})

# Verificar los cambios aplicados
df[bin_cols].head()
```

```python
df
```

```python
# Reemplazar "No phone service" por "No" en la columna MultipleLines
df['MultipleLines'] = df['MultipleLines'].replace('No phone service', 'No')

# Verificar los valores √∫nicos despu√©s del reemplazo
df['MultipleLines'].value_counts()
```

```python
# Transformar "Yes"/"No" a 1/0 en MultipleLines
df['MultipleLines'] = df['MultipleLines'].map({'Yes': 1, 'No': 0})

# Verificar el resultado
df[['MultipleLines']].head()
```

```python
# Reemplazar valores en la columna 'Contract'
df['Contract'] = df['Contract'].replace({
    'Month-to-month': '1',
    'One year': '2',
    'Two year': '3'
})

df['Contract'] = df['Contract'].astype('int64')

# Verificar que los cambios se hayan aplicado correctamente
df['Contract'].value_counts()
```

```python
# Columnas que se transformar√°n: "S√≠" ‚Üí 1, todo lo dem√°s ‚Üí 0
cols_to_binary = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
       'TechSupport', 'StreamingTV', 'StreamingMovies']

# Aplicar transformaci√≥n
for col in cols_to_binary:
    df[col] = df[col].apply(lambda x: 1 if x == 'Yes' else 0)

# Verificar los resultados
df
```

```python
#verificar si hay valores repetidos en la columna 'customerID'
df['customerID'].duplicated().sum()
df
```

```python
# Diccionario para abreviar las formas de pago
payment_map = {
    'Electronic check': '1',
    'Mailed check': '2',
    'Bank transfer (automatic)': '3',
    'Credit card (automatic)': '4'
}

# Aplicar la transformaci√≥n en la columna MetodoPago
df['PaymentMethod'] = df['PaymentMethod'].replace(payment_map)

# Verificar los cambios
df['PaymentMethod'].value_counts()
```

```python
# Crear un diccionario con los nuevos nombres de las columnas
column_rename_map = {
    'customerID': 'id_cliente',
    'Churn': 'evasion',
    'gender': 'genero',
    'SeniorCitizen': 'senior',
    'Partner': 'pareja',
    'Dependents': 'dependientes',
    'tenure': 'antiguedad',
    'PhoneService': 's_telefono',
    'MultipleLines': 'multipleslineas',
    'InternetService': 's_internet',
    'OnlineSecurity': 'seguridad_ol',
    'OnlineBackup': 'respaldo_ol',
    'DeviceProtection': 'proteccion_d',
    'TechSupport': 'serv_tec',
    'StreamingTV': 'tv_str',
    'StreamingMovies': 'pel_str',
    'Contract': 'contrato',
    'PaperlessBilling': 'boleta_elec',
    'PaymentMethod': 'met_pago',
    'Charges.Monthly': 'pago_mensual',
    'Charges.Total': 'pago_total',
    'Cuentas_Diarias': 'cuentas_diarias'
}

# Renombrar columnas en df
df.rename(columns=column_rename_map, inplace=True)

# Verificar los nuevos nombres de columnas
df.columns
```

```python
# Reemplazar 'Yes' por 'S√≠' en la columna 'evasion'
df['evasion'] = df['evasion'].replace({'Yes': '1'}).replace({'':'0'}).replace({'No': '0'}).replace({'Si': '1'})

df['evasion'] = df['evasion'].astype('int64')

# Verificar los cambios
df['evasion'].value_counts()
```

```python
# Reemplazar female y male en la columna genero
df['genero'] = df['genero'].replace({'Female': '1'}).replace({'Male':'0'})

df['genero'] = df['genero'].astype('int64')

# Verificar los cambios
df['genero'].value_counts()
```

```python
# Reemplazar servicios de internet en la columna s_internet
df['s_internet'] = df['s_internet'].replace({'Fiber optic': '1'}).replace({'DSL':'2'}).replace({'No':'0'})

df['s_internet'] = df['s_internet'].astype('int64')

# Verificar los cambios
df['s_internet'].value_counts()
```

```python
# Convertir pago_total a num√©rico
df['pago_total'] = pd.to_numeric(df['pago_total'], errors='coerce')

# Reemplazar valores faltantes usando antig√ºedad * pago_mensual
mask = df['pago_total'].isna()
df.loc[mask, 'pago_total'] = (
    df.loc[mask, 'antiguedad'] * df.loc[mask, 'pago_mensual']
)
```

```python
# cambiar tipo de dato de la columna contrato a int64
df['contrato'] = df['contrato'].astype('int64')

# Verificar los cambios
df['contrato'].value_counts()
```

```python
# cambiar tipo de dato de la columna met_pago a int64
df['met_pago'] = df['met_pago'].astype('int64')

# Verificar los cambios
df['met_pago'].value_counts()
```

```python
# Verificar que ya no hay valores NaN y que el tipo de dato es float64
df.info()
```

```python
# Columnas de servicios contratados (excluyendo s_internet)
servicios = df.columns[[7,8,10,11,12,13,14,15]]

# Asegurarse de que todas sean binarias
df[servicios] = df[servicios].applymap(lambda x: 1 if x != 0 else 0)

# Separar por evasi√≥n
evaden = df[df['evasion'] == 1]
no_evaden = df[df['evasion'] == 0]

# Calcular contrataciones por servicio
evaden_sums = evaden[servicios].sum()
no_evaden_sums = no_evaden[servicios].sum()

# Crear DataFrame para stacked bar
df_stacked = pd.DataFrame({
    'Servicio': servicios,
    'Evaden': evaden_sums.values,
    'No evaden': no_evaden_sums.values
})

# Convertir a formato largo
df_stacked = df_stacked.melt(id_vars='Servicio',
                             var_name='Grupo',
                             value_name='Cantidad')

# Calcular total por servicio para porcentaje
totales_por_servicio = df_stacked.groupby('Servicio')['Cantidad'].transform('sum')
df_stacked['Porcentaje'] = (df_stacked['Cantidad'] / totales_por_servicio * 100).round(2)

# Ordenar por total
orden = (evaden_sums + no_evaden_sums).sort_values(ascending=False).index.tolist()
df_stacked['Servicio'] = pd.Categorical(df_stacked['Servicio'], categories=orden, ordered=True)
df_stacked = df_stacked.sort_values('Servicio')

# Gr√°fico de barras apiladas (porcentaje)
fig = px.bar(df_stacked,
             x='Servicio', y='Porcentaje', color='Grupo',
             text='Porcentaje',
             title='Distribuci√≥n porcentual de servicios contratados: Evaden vs No evaden',
             color_discrete_map={'Evaden': 'orange', 'No evaden': 'lightblue'})

fig.update_traces(texttemplate='%{text}%', textposition='inside')
fig.update_layout(barmode='stack', xaxis_title='Servicio', yaxis_title='Porcentaje')

#ajustar tama√±o
fig.update_layout(width=800, height=500)

fig.show()

#guardar
fig.write_image("/content/grafico_evasion_servicios.png")
```

```python
# Columnas de servicios y convertir a binario
servicios = [
    's_telefono', 'multipleslineas', 's_internet', 'seguridad_ol',
    'respaldo_ol', 'proteccion_d', 'serv_tec', 'tv_str', 'pel_str'
]
df[servicios] = df[servicios].applymap(lambda x: 1 if x != 0 else 0)

# Variables independientes + categ√≥ricas convertidas
X = df[servicios + ['contrato', 'met_pago', 'genero', 'cuentas_diarias']]
X = pd.get_dummies(X, columns=['contrato', 'met_pago'], drop_first=True)
X = sm.add_constant(X)
X = X.astype('float64')

# Variable dependiente
y = df['evasion'].astype('float64')

# Alinear X e y en caso de valores faltantes
X = X.dropna()
y = y.loc[X.index]

# Ajustar modelo
modelo = sm.OLS(y, X).fit()

# Preparar para graficar coeficientes
coef_df = modelo.params.reset_index()
coef_df.columns = ['Variable', 'Coeficiente']
coef_df = coef_df[coef_df['Variable'] != 'const']
coef_df['Comparado con'] = coef_df['Variable'].apply(
    lambda v: 'Contrato mensual' if 'contrato_' in v else
              'Cheque electr√≥nico' if 'met_pago_' in v else
              'Referencia impl√≠cita' if v == 'genero' else
              'Base' if v in servicios else ''
)

# Ordenar por impacto
coef_df_sorted = coef_df.sort_values(by='Coeficiente', key=abs, ascending=False)

# Graficar
plt.figure(figsize=(10, 8))
sns.barplot(data=coef_df_sorted, x='Coeficiente', y='Variable', palette='coolwarm')
plt.title('Impacto de variables sobre la evasi√≥n\n(Comparado con contrato mensual y cheque electr√≥nico)')
plt.axvline(0, color='gray', linestyle='--')
plt.tight_layout()

coef_df_sorted[['Variable', 'Coeficiente', 'Comparado con']].round(3)

#guardar
plt.savefig("/content/grafico_impacto_variables.png")

plt.show()
```
